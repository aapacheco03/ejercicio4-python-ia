````python
import os
import gradio as gr
from google import genai
from transformers import pipeline

# --- Configuration & Initialization ---
GEMINI_API_KEY = os.environ.get("GEMINI_API_KEY")

if not GEMINI_API_KEY:
    try:
        from google.colab import userdata
        GEMINI_API_KEY = userdata.get('GEMINI_API_KEY')
    except Exception:
        pass

if GEMINI_API_KEY:
    client = genai.Client(api_key=GEMINI_API_KEY)
    # Using Gemini 2.0 Flash for optimal performance and quota availability
    GEMINI_MODEL = "gemini-2.0-flash"
    print(f"✅ Gemini API Key detected. Using model: {GEMINI_MODEL}")
else:
    client = None
    print("Warning: GEMINI_API_KEY not found. Generative features will be disabled.")

try:
    sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
except Exception as e:
    sentiment_analyzer = None
    print(f"Error loading sentiment model: {e}")

def analyze_sentiment(text):
    if not text.strip():
        return "Please provide feedback text.", "", ""
    if not sentiment_analyzer:
        return "Sentiment model not available.", "", ""
    try:
        result = sentiment_analyzer(text)[0]
        label = result['label']
        score = f"{result['score']:.2%}"
        interpretation = "Positive sentiment indicates stable morale. Maintain current engagement strategies." if label == "POSITIVE" else "Negative sentiment detected. This suggests operational friction or risk. Immediate investigation or listening sessions recommended."
        return label, score, interpretation
    except Exception as e:
        return f"Error: {str(e)}", "", ""

def generate_executive_summary(metrics):
    if not metrics.strip(): return "Please provide operational metrics."
    if not client: return "Gemini API key not configured."
    prompt = f"As a WFM Expert, analyze these metrics and provide a structured report with: 1. Executive Summary 2. Key Risks 3. Recommended Actions 4. Leadership-Ready Message. Metrics: {metrics}"
    try:
        response = client.models.generate_content(model=GEMINI_MODEL, contents=prompt)
        return response.text
    except Exception as e:
        return f"Error generating summary: {str(e)}"

def draft_communication(situation):
    if not situation.strip(): return "Please provide the operational situation."
    if not client: return "Gemini API key not configured."
    prompt = f"Draft a professional operational communication based on this situation. Provide: 1. Subject Line 2. Professional Message 3. Action Required 4. Tone Check description. Situation: {situation}"
    try:
        response = client.models.generate_content(model=GEMINI_MODEL, contents=prompt)
        return response.text
    except Exception as e:
        return f"Error drafting communication: {str(e)}"

# Main UI Application
with gr.Blocks(theme=gr.themes.Soft(), title="WFM Copilot") as demo:
    gr.Markdown("# 🛠️ WFM Copilot")
    gr.Markdown("Transforming operational data into actionable insights and professional communications.")
    with gr.Tabs():
        with gr.TabItem("📊 Feedback Sentiment Analyzer"):
            gr.Markdown("### Analyze workforce feedback sentiment")
            sentiment_input = gr.Textbox(label="Agent/Stakeholder Feedback", placeholder="Enter feedback here...", lines=3)
            sentiment_btn = gr.Button("Analyze Sentiment", variant="primary")
            with gr.Row():
                sentiment_label = gr.Textbox(label="Sentiment Label")
                sentiment_score = gr.Textbox(label="Confidence Score")
            sentiment_interpret = gr.Textbox(label="Operational Interpretation", lines=3)
            sentiment_btn.click(analyze_sentiment, inputs=sentiment_input, outputs=[sentiment_label, sentiment_score, sentiment_interpret])
        with gr.TabItem("📝 Executive Summary Generator"):
            gr.Markdown("### Generate summaries from operational metrics")
            summary_input = gr.Textbox(label="Operational Metrics / Notes", placeholder="e.g., SLA dropped from 83% to 72%.", lines=4)
            summary_btn = gr.Button("Generate Report", variant="primary")
            summary_output = gr.Markdown(label="Executive Report")
            summary_btn.click(generate_executive_summary, inputs=summary_input, outputs=summary_output)
        with gr.TabItem("📧 Operations Communication Draft"):
            gr.Markdown("### Draft professional team communications")
            comm_input = gr.Textbox(label="Context / Situation", placeholder="e.g., Notify agents about schedule adjustments.", lines=4)
            comm_btn = gr.Button("Draft Message", variant="primary")
            comm_output = gr.Markdown(label="Communication Draft")
            comm_btn.click(draft_communication, inputs=comm_input, outputs=comm_output)

if __name__ == "__main__":
    demo.launch(debug=True)
````
